```python
data=[
    {"name":"Jahidule", "Age":25, "salary":25000},
     {"name":"Rahim", "Age":None, "salary":30000},
     {"name":"Karim", "Age":25, "salary":40000},
     {"name":"Hamidule", "Age":None, "salary":50000},
     {"name":"Jishan", "Age":28, "salary":None},
     {"name":"Tom", "Age":25, "salary":23000},
     {"name":"Jonh", "Age":30, "salary":12000},
     {"name":"Kim", "Age":32, "salary":None},
     {"name":"Jak", "Age":21, "salary":25000},
     {"name":"Sara", "Age":None, "salary":25000},
     {"name":"Carina", "Age":19, "salary":None}

]
```

```python
print("Original Data")
for row in data:
  print(row)
```

```python
print("Detect Missing Value (None)")
for i, row in enumerate(data):
  for key, value in row.items():
    if value is None:
      print(f"Row {i+1}, column'{key}' value is None")
```

```python
cleaned_data=[]
for row in data:
  if None not in row.values():
    cleaned_data.append(row)
print("After Removig Row After Missing Value: ")
for row in cleaned_data:
    print(row)
```

```python
ages = [row["Age"] for row in data if row["Age"] is not None]
salaries = [row["salary"] for row in data if row["salary"] is not None]
```

```python
mean_age = sum(ages) / len(ages)
mean_salary = sum(salaries) / len(salaries)
```

```python
for row in data:
    if row["Age"] is None:
        row["Age"] = round(mean_age, 1)
    if row["salary"] is None:
        row["salary"] = round(mean_salary, 1)

print("\nAfter Filling Missing Values with Mean:")
for row in data:
    print(row)
```

```python
filled_forward = []
prev_row = None

for row in data:
    new_row = row.copy()
    if prev_row:
        for key, value in new_row.items():
            if value is None:
                new_row[key] = prev_row[key]
    filled_forward.append(new_row)
    prev_row = new_row

print("\n After Forward Fill (Manual Implementation):")
for row in filled_forward:
    print(row)
```
