```python
number_list = [1, 2, 3, 3, 4, 4, 5, 6, 7, 8, 9]

odd_list = list()
even_list = list()

for number in number_list:
    if number%2==0:
        even_list.append(number)
    else:
        odd_list.append(number)

class Stat:
    def __init__(self, data):
        self.data = data

    def mean(self):
        return sum(self.data)/len(self.data)

    # data_sort > median: Middle position data > len-> odd : len/2; len->even : ([len/2] + [len/2 - 1])/2
    def median(self):
        sorted_data = sorted(self.data)
        if len(sorted_data)%2==1:
            return sorted_data[len(sorted_data)//2]
        else:
            return (sorted_data[len(sorted_data)//2] + sorted_data[len(sorted_data)//2 -1])/2

    # Count each element frequency -> maximum frequency element
    def mode(self):
        frequency = dict()

        #frequency = {1:1, 3:2,}
        for number in self.data:
            if number not in frequency.keys():
                frequency[number] = 1
            else:
                frequency[number] += 1

        max_frequency = 0

        for number in frequency.keys():
            if frequency[number] > max_frequency:
                max_frequency = frequency[number]
                calculated_mode = number

        return calculated_mode

    def display(self):
        print(f"Data: {self.data}")
        print("--------------Stat of above data---------------------")
        print(f"Mean: {self.mean()}")
        print(f"Mode: {self.mode()}")
        print(f"Median: {self.median()}")
        print("--------------End---------------------")



odd_stat = Stat(odd_list)
even_stat = Stat(even_list)

odd_stat.display()
```
