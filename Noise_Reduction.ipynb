```python
data = [4, 8, 15, 21, 21, 24, 25, 28, 34, 36, 40, 42]
```

```python
print("ðŸ”¹ Original Data:")
print(data)
```

```python
sorted_data=sorted(data)
print("ðŸ”¹ Sorted Data:")
print(sorted_data)
```

```python
bin_size = 4
bins = []
```

```python
for i in range(0, len(sorted_data), bin_size):
    bins.append(sorted_data[i:i+bin_size])

print("\n Bins (Before Smoothing):")
for b in bins:
    print(b)
```

```python
bins_mean = []
for b in bins:
    mean_value = sum(b) / len(b)
    bins_mean.append([round(mean_value, 2)] * len(b))  # replace all with mean

print("\n Binning by Mean:")
for b in bins_mean:
    print(b)
```

```python
bins_median = []
for b in bins:
    n = len(b)
    if n % 2 == 0:
        median = (b[n//2 - 1] + b[n//2]) / 2
    else:
        median = b[n//2]
    bins_median.append([median] * len(b))

print("\n Binning by Median:")
for b in bins_median:
    print(b)
```

```python
bins_boundary = []
for b in bins:
    lower, upper = b[0], b[-1]
    new_bin = []
    for val in b:
        # Replace with closest boundary
        if abs(val - lower) <= abs(val - upper):
            new_bin.append(lower)
        else:
            new_bin.append(upper)
    bins_boundary.append(new_bin)

print("\n Binning by Boundary:")
for b in bins_boundary:
    print(b)
```
