```python
data = [50, 60, 70, 80, 90, 100]

print("ðŸ”¹ Original Data:")
print(data)
```

```python
def min_max_normalization(data):
    min_val = min(data)
    max_val = max(data)
    normalized = []
    for x in data:
        new_val = (x - min_val) / (max_val - min_val)
        normalized.append(round(new_val, 3))
    return normalized

min_max_result = min_max_normalization(data)
print("\n Min-Max Normalization (0â€“1 Range):")
print(min_max_result)
```

```python
def z_score_normalization(data):
    n = len(data)
    mean = sum(data) / n

    variance = sum((x - mean) ** 2 for x in data) / n
    std_dev = variance ** 0.5

    z_scores = []
    for x in data:
        z = (x - mean) / std_dev
        z_scores.append(round(z, 3))
    return z_scores, round(mean, 3), round(std_dev, 3)

z_score_result, mean, std_dev = z_score_normalization(data)

print("\n Z-Score Normalization:")
print(z_score_result)
print(f"Mean = {mean}, Standard Deviation = {std_dev}")
```
